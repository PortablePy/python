### Output Rendering Methods. 

Output can rendered in many methods in pygal. 

Render pyquery will be covered in next demo. 

```python
import pygal 
import pandas as pd
import os
import webbrowser
```

```python
tokyo_col = [84.78, 36.41, 61.89, 81.72, 54.41, 113.09]
oslo_col = [107.14, 48.58, 79.42, 93.81, 114.42, 104.3]
nairobi_col = [41.76, 13.17, 28.23, 36.97, 35.98, 30.62]
```

```python
from pygal.style import BlueStyle
```

```python
radar_chart = pygal.Radar(width = 800,
                          height = 450,
                          explicit_size = True,
                          style=BlueStyle)

radar_chart.title = 'Cost Of Living Index'

radar_chart.x_labels = ['COL ex. Rent,', 'Rent', 'COL + Rent', 
                        'Groceries', 'Restaurants', 'Local PP']
```

```python
radar_chart.add('Tokyo', tokyo_col)
radar_chart.add('Oslo', oslo_col)
radar_chart.add('Nairobi', nairobi_col)
```

### Render 

The ouput can be render as **"SVG" bytes**. It can either display it or save it to a file.

```python
radar_chart.render()
```

### Use the SVG bytes to write to a file.

```python
file_data = radar_chart.render()
```

```python
with open('radar_output_file.svg','wb') as file:
    file.write(file_data)
```

You can open that file in the browser by using **OS** module (find the absolute path)
**define a method to open in the browser**

```python
def open_in_brower(file_to_open, browser):
    
    file_path = os.path.realpath(file_to_open)
    file_path = 'file://' + file_path
    browser = webbrowser.get(browser)
    browser.open(file_path)
```

```python
open_in_brower('radar_output_file.svg', 'chrome')
```

```python
open_in_brower('radar_output_file.svg', 'safari')
```

```python
open_in_brower('radar_output_file.svg', 'firefox')
```

## Base64

A format that is extensively used to send images over the network. When you want to send you image in JSON file using API then you can convert your image to **base64** encoding. 

Coming demo - This method will be used. In the flask demo. 

```python
base_64_image  = radar_chart.render_data_uri()
```

**render_data_uri** method  adds in the Content-Type (format type) header at the front of the "Base64" text. It must has a specific purpose we will see over coming demos. For now, it should be removed to decode the "Base64" text. You can also see that it's a **SVG** image converted to base64 format. 

```python
base_64_image 
```

Remove the **'Content-Type'** header to convert back to **SVG file** bytes. 

```python
import base64

imgdata = base64.b64decode(base_64_image.split(',')[1])

imgdata
```

**Since the data is in the form of bytes. Let's write to a file.**

```python
with open('base_64_decoded_image.svg','wb') as file:
    file.write(imgdata)
```

**Display the SVG file using the IPython Display module although the image is not interactive. We can use the "display_chart" code to make the chart interactive in the Jupyter notebook.**

```python
from IPython.display import display, SVG

display(SVG('base_64_decoded_image.svg'))
```

## Render Tree

#### Rendered tree is a XML format that is rendered to be used at server side applications. For e.g. used with django. 

```python
rendered_tree_file = radar_chart.render_tree()
```

```python
type(rendered_tree_file)
```

**Attributes of the tree.**

```python
rendered_tree_file.attrib
```

##### Let's print out the all the children.

```python
for elem in rendered_tree_file.iterchildren():
    print (elem.tag, elem.text)
```

It is a SVG file in the form of xml file. Tags give you information about how the file was generated.

**To printout the whole content of the tree. **  It is of **"bytes form"** which can be written to a **XML** file.

```python
from lxml import etree

print(etree.tostring(rendered_tree_file, 
                     pretty_print=True))
```

```python
tree_data = etree.tostring(rendered_tree_file)
```

**Write the data to file ** You can open the XML file using the browser to show the interactive chart.

```python
with open('etree_data.xml', 'wb') as f:
    f.write(tree_data)
```

#### Open the file in the browser

```python
open_in_brower('etree_data.xml', 'chrome')
```
