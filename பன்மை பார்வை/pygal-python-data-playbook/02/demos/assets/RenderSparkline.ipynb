## Sparklines

**Sparkline is small graphic designed to give a quick representation of numerical or statistical information using small graph without axes.**

We will use the line_chart from one of the previous demos to sparkline text

**Note**: Rendering sparkline takes very long time. The numbers that are being are cut down. So, the stock data being plotted is no more than a week.


```python
import pygal
from IPython.display import display, HTML, SVG
```

Get the data to plot.

```python
stock_closing_price = [165.48, 163.94, 166.07, 160.89,
                       156.83, 150.73, 146.83]
```

##### Initialize the pygal object and add in the data. 

```python
line_chart = pygal.Line(width = 640,
                        height = 360,
                        explicit_size = True)

line_chart.add('Apple', 
               stock_closing_price)
```

##### Display the sparkline with inbuilt Ipython display method. 

```python
display(SVG(line_chart.render_sparkline()))
```

#### It is evident that stock price has gone down. The purpose of this sparkline is fulfilled as it gives you a quick outline of the data using a small graph. 

##### Styles can be added to the sparkline using themes in the pygal catalog. 

```python
from pygal.style import DarkSolarizedStyle

line_chart = pygal.Line(width = 640,
                        height = 360,
                        explicit_size = True,
                        style=DarkSolarizedStyle)
```

```python
line_chart.x_labels = map(str, range(1, len(stock_closing_price)))

line_chart.add('Apple', 
               stock_closing_price)
```

#### You can also define options in the while rendering. We will define height and width of the graph. 



```python
display(SVG(line_chart.render_sparkline(width=500, 
                                        height=200, 
                                        show_dots = True)))
```

#### You can show the XY label in sparkline. 

```python
display(SVG(line_chart.render_sparkline(width=500, 
                                        height=200,
                                        show_x_labels = True, 
                                        show_y_labels = True)))
```

### You can render a sparktext using pygal. 

It similar to small bar graph.

```python
line_chart.render_sparktext()
```

You can specify the explicit minimum for the values.

First, sparktext is relative to minimum value i.e 146.83. Here the values to are relative to 130.

```python
line_chart.render_sparktext(relative_to = 130)
```
