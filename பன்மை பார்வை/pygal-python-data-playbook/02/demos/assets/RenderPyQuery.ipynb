## Pyquery 

You can render your chart in pyquery format. 

**Pyquery is a jquery in python where you can query the tags and manipulate the tags (data) in side the XML or HTML file.**

Pygal returns a pyquery objects of the svg charts 

#### In the pygal sources, it does the following 

**PyQuery(chart.render( ), parser='html')** 

It is evident that pygal is rendering the  **SVG file** to html so that it can be used to query.

Import the statements as required.

```python
import pygal 
import pandas as pd
from pygal.style import BlueStyle
from pyquery import PyQuery as pq
from pprint import pprint
```

Here, the chart from previous demo is used.

```python
tokyo_col = [84.78, 36.41, 61.89, 81.72, 54.41, 113.09]
oslo_col = [107.14, 48.58, 79.42, 93.81, 114.42, 104.3]
nairobi_col = [41.76, 13.17, 28.23, 36.97, 35.98, 30.62]

radar_chart = pygal.Radar(width = 600,
                          height = 400,
                          explicit_size = True,
                          style=BlueStyle)
radar_chart.title = 'Cost Of Living Index'

radar_chart.x_labels = ['COL ex. Rent,', 'Rent', 'COL + Rent', 
                        'Groceries', 'Restaurants', 'Local PP']

radar_chart.add('Tokyo', tokyo_col)
radar_chart.add('Oslo', oslo_col)
radar_chart.add('Nairobi', nairobi_col)
```

Let's render the data to pyquert object.

```python
pyq_object = radar_chart.render_pyquery()
```
